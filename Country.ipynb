```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('data.csv')
print(dataset)
```

```python
def RegressaoLinear(x,y):   
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
    model = LinearRegression()
    model.fit(x_train, y_train)

    y_pred = model.predict(x_test)
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)

    print(f"Mean Squared Error (MSE): {mse}")
    print(f"R-squared (R2): {r2}")

    plt.figure(figsize=(12, 5))
    # Plotar os dados reais e as previsões
    plt.scatter(x_test, y_test, color='red', label='Dados Reais', s = 15)
    plt.plot(x_test, y_pred, color='blue', linewidth=3, label='Previsões')
    plt.xlabel('Previous Scores')
    plt.ylabel('Performance Index')
    plt.legend()
    plt.show()
```

```python
RegressaoLinear(dataset[['Governance']],dataset['AveragScore'])
```

Podemos ver que existe uma relação entre a qualidade da democracia, estado de direito e eficácia do governo e a pontuação média geral do país em todos os indicadores.

```python
RegressaoLinear(dataset[['EconomicQuality']],dataset['Education'])
```
